Matplotlib是一个强大的数据可视化库,可用于创建各种静态、动态和交互式的图表。
Matplotlib is a powerful data visualization library that can be used to create a wide variety of static, animated, and interactive visualizations.

我们将学习如何使用Matplotlib创建各种类型的图表,并对图表进行自定义和美化。
we will learn how to use Matplotlib to create various types of plots, and customize and beautify the plots.

```python
import numpy as np
from matplotlib import pyplot as plt
```

生成100个从0到2π的等间距数字,计算这些数字的正弦值, 绘制图表.
Generate 100 evenly spaced numbers from 0 to 2π. Calculate the sine of these numbers, Plot the graph

```python
x = np.linspace(0, 2 * np.pi, 100)  # 生成100个从0到2π的等间距数字 | Generate 100 evenly spaced numbers from 0 to 2π
y = np.sin(x)  # 计算这些数字的正弦值 | Calculate the sine of these numbers

plt.plot(x, y)
```

```python
plt.show()  # 显示图表 | Display the plot
# plt.savefig('sin_function.png')  # 保存图表为PNG文件 | Save the plot as a PNG file
```

## 图表的标题和标签 | Plot Title and Labels

我们可以使用title()、xlabel()和ylabel()函数为图表添加标题和轴标签。
We can use the title(), xlabel(), and ylabel() functions to add a title and axis labels to the plot.

```python
plt.plot(x, y)
plt.title('Sine Function')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

## 图表的样式自定义 | Plot Style Customization

我们可以使用各种参数来自定义图表的样式,如线条颜色、线条样式、标记样式等。
We can use various parameters to customize the style of the plot, such as line color, line style, marker style, etc.

```python
plt.figure(figsize=(8, 6), dpi=80)  # 设置图表的大小和分辨率 | Set the size and resolution of the figure
plt.plot(x, y, color='red', linewidth=2, linestyle='--', marker='o', markersize=5)  # 设置线条和标记的样式 | Set the style of the line and markers
plt.title('Sine Function', fontsize=18)  # 设置标题和字体大小 | Set the title and font size
plt.xlabel('x', fontsize=14)
plt.ylabel('sin(x)', fontsize=14)
plt.xticks(fontsize=12)  # 设置x轴刻度的字体大小 | Set the font size of x-axis ticks
plt.yticks(fontsize=12)
plt.grid(True)  # 显示网格 | Show grid
plt.show()
```


## 绘制多条曲线 | Plotting Multiple Curves

我们可以在同一个图表上绘制多条曲线,并使用legend()函数添加图例。
We can plot multiple curves on the same plot, and use the legend() function to add a legend.


```python
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.title('Sine and Cosine Functions')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()  # 添加图例 | Add legend
plt.show()
```

## 绘制散点图 | Plotting Scatter Plots

我们可以使用scatter()函数绘制散点图。
We can use the scatter() function to plot scatter plots.

```python
x = np.random.rand(100)
y = np.random.rand(100)
colors = np.random.rand(100)  # 随机生成每个点的颜色 | Randomly generate colors for each point
sizes = 1000 * np.random.rand(100)  # 随机生成每个点的大小 | Randomly generate sizes for each point

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5)  # 设置点的颜色、大小和透明度 | Set the color, size, and transparency of the points
plt.title('Random Scatter Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

## 绘制条形图 | Plotting Bar Plots

我们可以使用bar()函数绘制垂直条形图,使用barh()函数绘制水平条形图。
We can use the bar() function to plot vertical bar plots, and the barh() function to plot horizontal bar plots.

```python
x = ['A', 'B', 'C', 'D', 'E']
y = [3, 7, 2, 5, 1]

plt.bar(x, y)
plt.title('Bar Plot')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

## 绘制直方图 | Plotting Histograms

我们可以使用hist()函数绘制直方图,显示数据的分布情况。
We can use the hist() function to plot histograms, showing the distribution of the data.

```python
data = np.random.normal(0, 1, 1000)  # 随机生成1000个服从标准正态分布的数据 | Randomly generate 1000 data points from a standard normal distribution

plt.hist(data, bins=30)  # 设置直方图的条块数 | Set the number of bins in the histogram
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```


## 绘制三维图 | Plotting 3D Graphs

我们可以使用Axes3D模块绘制三维图。
We can use the Axes3D module to plot 3D graphs.


```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

ax.plot_surface(X, Y, Z)
plt.title('3D Surface Plot')
plt.show()
```
